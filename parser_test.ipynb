```python
import json
import re

def recover_json(raw: str):

    # 2️⃣ \" → "  (JSON escape 정상화)
    raw = raw.replace(r'\"', '"')

    # 3️⃣ \\n → \n
    raw = raw.replace(r'\\n', '\n')

    # 4️⃣ unicode escape 해제 (\u003c → <)
    raw = raw.encode().decode("unicode_escape")

    # 5️⃣ 이제 진짜 JSON 상태 → dict 변환
    data = json.loads(raw)

    return data
```

```python
import requests
import re
import json
from bs4 import BeautifulSoup

def get_sqld_exam(num:int):
    url = f"https://sqlwise.com/ko/sqld/bank/{num}"
    
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    
    html = requests.get(url, headers=headers).text
    
    # 문제 데이터 블록 찾기
    
    start_str = '{\\"exam\\"'
    end_str = '\\"difficulty\\":'
    
    start_idx = html.find(start_str)
    end_idx = html.find(end_str)
    
    exam_data = json.loads(html[start_idx:end_idx + len(end_str) + 2].replace('\\', '').replace('u003c/pu003en', '').replace('u003cpu003e',''))
    return exam_data
```

```python
import time
from tqdm.notebook import tqdm

exam_list = []
failed_list = []

for num in tqdm(range(1, 442)):
    try:
        exam_list.append(get_sqld_exam(num))
    except Exception as e:
        failed_list.append(num)
        print(f'{num}번 문제 추출 실패')
    time.sleep(4)
```
