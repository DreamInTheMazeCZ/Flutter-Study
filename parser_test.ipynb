```python
from bs4 import BeautifulSoup
import requests, re, json

def get_sqld_exam(num:int):
    url = f"https://sqlwise.com/ko/sqld/bank/{num}"

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    html = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html, 'html.parser')

    # 문제 데이터 JSON 부분 추출
    start_str = '{\\"exam\\"'
    end_str = '\\"difficulty\\":'

    start_idx = html.find(start_str)
    end_idx = html.find(end_str)

    processing = html[start_idx:end_idx + len(end_str) + 2]

    # 각종 특수문자 제거 및 JSON 파싱
    text = re.sub(r'\\u[0-9a-fA-F]{4}', '', processing)
    text = re.sub(r'\\n', '', text)
    text = re.sub(r'/p', '', text)
    text = text.replace('\\', '')

    try:
        data = json.loads(text)
    except Exception as e:
        # context 데이터 처리
        start_con = '"context":"'
        end_con = ',"erdata":'

        con_start_idx = text.find(start_con)
        con_end_idx = text.find(end_con)

        context_text = text[con_start_idx:con_end_idx]
        text = text.replace(context_text, '').replace(',,', ',')

        data = json.loads(text)
        
        # code 태그 내 문제 추출
        con_data = [i.text for i in soup.find_all('code')][0]
        data = {**data, 'context': con_data}

    return data
```

```python
import time
from tqdm.notebook import tqdm

exam_list = []
failed_list = []

for num in tqdm(range(1, 442)):
    try:
        exam_list.append(get_sqld_exam(num))
    except Exception as e:
        failed_list.append((num, str(e)))
        print(f'{num}번 문제 추출 실패')
    time.sleep(1)
```
